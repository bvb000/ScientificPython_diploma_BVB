#### **Exercise 1:**
Create a function returning the first N terms of the tailor series of the function without using the sympy series function.

```python
import sympy as sp
import numpy as np
from numpy import random

x, y, z = sp.symbols('x y z')
sp.init_printing(use_unicode=True)

def fxn(x):
    return sp.sin(x)
    
N = random.randint(10)
a = 2*sp.pi*random.sample()

print("N=",N)
print("a=",a)

y=fxn(x)
saver=y
for i in range(1,N):
    saver=sp.diff(saver,x)
    y=y+saver.subs(x,a)*(x-a)**i/sp.factorial(i)

y
```

#### **Exercise 2:**

Write a matrix multiplication function using lists ("manually", using the matrix multiplication formula) and another one using numpy arrays (call matmul function). Time them both. 

```python
%%timeit
import numpy as np
from numpy import random

N = random.randint(10)

mat1=[[1]*N]*N
mat2=[[1]*N]*N
product=[[1]*N]*N

for i in range (0,N):
    for j in range (0,N):
        mat1[i][j]=random.sample()
        mat2[i][j]=random.sample()

for i in range (0,N):
    for j in range (0,N):
        for k in range (0,N):
            product[i][j]=product[i][j]+mat1[i][k]*mat2[k][j]
```

#### **Exercise 4:**
Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
import numpy as np
from matplotlib import pyplot as plt

recaman = [0]

N=int(input("please state N"))

for i in range (1,N):
    if ((recaman[i-1]-i)>0) and (recaman.count(recaman[i-1]-i)==0):
        recaman.append(recaman[i-1]-i)
    else:
        recaman.append(recaman[i-1]+i)

x = np.linspace(-50,50,500)
fig = plt.figure(figsize=(14,7))
        
for i in range (1,N):
    if (recaman[i]>recaman[i-1]):
        if (recaman[i]%2==0):
            y=(((recaman[i]-recaman[i-1])/2)**2-(x-(recaman[i]+recaman[i-1])/2)**2)**0.5
            plt.plot(x,y,color='r')
        else:
            y=-1.*(((recaman[i]-recaman[i-1])/2)**2-(x-(recaman[i]+recaman[i-1])/2)**2)**0.5
            plt.plot(x,y,color='r')
        
print(recaman)
```

#### **Exercise 6:**

Plot a nice-looking 3-D torus. Do this 3 times with different styles. 

```python
from matplotlib import cm

u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:50j]

R=10
r=2
x = (R+r*np.sin(v))*np.cos(u)
y = (R+r*np.sin(v))*np.sin(u)
z = r*np.cos(v)

fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.set_box_aspect([1+R/r, 1+R/r, 1])
ax.plot_surface(x, y, z, cmap=cm.rainbow)

fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.set_box_aspect([1+R/r, 1+R/r, 1])
ax.plot_wireframe(x, y, z)

fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.set_box_aspect([1+R/r, 1+R/r, 1])
ax.scatter3D(x, y, z, cmap=cm.rainbow)
```

#### **Exercise 7:**

Repeat the curve fitting example (from SciPy lecture) with the following modifications:

 - choose some other function
 - first write the data to a file, then read it (with numpy, for example)
 - produce the picture that has axes, name and legend


```python
from scipy import optimize

np.random.seed(0)

# Let's generate fake data that we can "fit" then 
def f(x, a, phi):
    return np.sin(a **2* x + phi)

#random makes it look like experimental data
x = np.linspace(0, 4, 50)
y = f(x, 1.3, 0.9) + .1*np.random.normal(size=50)

# Now fit the model: the parameters omega and phi can be found in the
# `params` vector
params, params_cov = optimize.curve_fit(f, x, y)

print(params)

# plot the data and the fitted curve
plt.plot(x, y, 'bx')
plt.plot(x,f(x,1.3,0.9),'g-')
plt.plot(x, f(x, *params), 'r-')
```
